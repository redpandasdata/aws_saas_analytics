```python
def date_format(df):
    df["Order Date"]= pd.to_datetime(df["Order Date"])
    df["month_name"] = df["Order Date"].dt.month_name()
    df["month"] = df["Order Date"].dt.month
    df["year"] = df["Order Date"].dt.year
    df["day_name"] = df["Order Date"].dt.day_name()
    df["weekday"] = df["Order Date"].dt.weekday
    df["day"] = df["Order Date"].dt.day
    df["quarter"] = df["Order Date"].dt.quarter

    return df

date_format(df)
```

```python
print(df["Order Date"].min())
print(df["Order Date"].max())
```

```python
df_sales_by_company = df.groupby("Customer")["Sales"].sum().round(0).reset_index().sort_values(by="Sales",ascending=False)

fig = px.bar(df_sales_by_company, x="Customer", y="Sales", barmode="group", title="Sales by company")
fig.show()
```

```python
df_sales_by_months = df.groupby("month")["Sales"].sum().round().reset_index().sort_values(by="month")

fig = px.line(df_sales_by_months, x="month", y='Sales', title='Sales by months')
fig.update_layout(xaxis={'dtick':1})
fig.show()
```
